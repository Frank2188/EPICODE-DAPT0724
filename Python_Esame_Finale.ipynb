```python
#SELEZIONE COLONNE D'INTERESSE
df_filtered = df[['continent','location','date', 'total_cases']]
# ORDINAMENTO PER DATA DISCENDENTE 
df_filtered = df_filtered.sort_values(by='date', ascending=False)
print(df_filtered) 
```

```python
df_cases_cleaned = df_filtered[df_filtered['total_cases'] > 0]
print (df_cases_cleaned.head(20)) # CI SONO DEI NAN TRA I CONTINENTI
```

```python
df_cases_with_nan = df_cases_cleaned[df_cases_cleaned.isna().any(axis=1)]
print(df_cases_with_nan.head(12)) # RAPPRESENTANO IL CONTEGGIO PER MACROREGIONI O CONTINENTI 
                                  # PRIMA DI DARLI PER BUONI FACCIO UN CONTEGGIO ANCHE PER NAZIONI
```

```python
"""PER VERIFICARE CHE IL CONTEGGIO PER CONTINENTE FOSSE CORRETTO
        FILTRO PER CONTINENTE """
df_europe = df_filtered[df_filtered['continent'].str.contains('Europe', case=False, na=False)]

""" E POI FILTRO PER ULTIMA DATA DI CONTEGGIO E VALORI COERENTI """
df_europe_cases = df_europe[(df_europe['total_cases'] > 0) & (df_europe['date'] == '2024-08-04')]
print(df_europe_cases)
```

```python
total_europe_sum = df_europe_cases['total_cases'].sum()

# FACCIO LA STESSA COSA PER GLI ALTRI CONTINENTI

df_oceania = df_filtered[df_filtered['continent'].str.contains('Oceania', case=False, na=False)]
df_oceania_cases = df_oceania[(df_oceania['total_cases'] > 0) & (df_oceania['date'] == '2024-08-04')]
total_oceania_sum = df_oceania_cases['total_cases'].sum()

df_asia = df_filtered[df_filtered['continent'].str.contains('Asia', case=False, na=False)]
df_asia_cases = df_asia[(df_asia['total_cases'] > 0) & (df_asia['date'] == '2024-08-04')]
total_asia_sum = df_asia_cases['total_cases'].sum()

df_africa = df_filtered[df_filtered['continent'].str.contains('Africa', case=False, na=False)]
df_africa_cases = df_africa[(df_africa['total_cases'] > 0) & (df_africa['date'] == '2024-08-04')]
total_africa_sum = df_africa_cases['total_cases'].sum()

df_northamerica = df_filtered[df_filtered['continent'].str.contains('North America', case=False, na=False)]
df_northamerica_cases = df_northamerica[(df_northamerica['total_cases'] > 0) & (df_northamerica['date'] == '2024-08-04')]
total_northamerica_sum = df_northamerica_cases['total_cases'].sum()

df_southamerica = df_filtered[df_filtered['continent'].str.contains('South America', case=False, na=False)]
df_southamerica_cases = df_southamerica[(df_southamerica['total_cases'] > 0) & (df_southamerica['date'] == '2024-08-04')]
total_southamerica_sum = df_southamerica_cases['total_cases'].sum()



print(f"La somma totale dei casi per l'Europa al 2024-08-04 è:{total_europe_sum:,.0f}")
print(f"La somma totale dei casi per l'Asia al 2024-08-04 è: {total_asia_sum:,.0f}")
print(f"La somma totale dei casi per l'Africa al 2024-08-04 è:{total_africa_sum:,.0f}")
print(f"La somma totale dei casi per l'North America al 2024-08-04 è:{total_northamerica_sum:,.0f}")
print(f"La somma totale dei casi per l'South America al 2024-08-04 è:{total_southamerica_sum:,.0f}")
print(f"La somma totale dei casi per l'Oceania al 2024-08-04 è:{total_oceania_sum:,.0f}")
```

```python
import pandas as pd
```

```python
df = pd.read_csv(url) 
```

```python
# SELEZIONE COLONNE D'INTERESSE
df_filtered = df[['location', 'date', 'total_cases', 'new_cases']].copy()
# TRASFORMAZIONE DATE IN FORMATO DATETIME
df_filtered['date'] = pd.to_datetime(df_filtered['date'])
# FILTRO PER NAZIONE E DATA
df_italy_2022 = df_filtered[(df_filtered['location'] == 'Italy') & (df_filtered['date'].dt.year == 2022)]
# RIMOZIONE DUPLICATI E VALORI = 0 PER VISUALIZZARE SOLO I GIORNI DI AGGIORNAMENTO DATI
df_italy_2022_unique = df_italy_2022.drop_duplicates(subset='total_cases', keep='first')
df_italy_2022_unique = df_italy_2022_unique[df_italy_2022_unique['new_cases'] > 0]
print(df_italy_2022_unique)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.dates as mdates

plt.figure(figsize=(16,7))
sns.lineplot(x='date', y='total_cases', data=df_italy_2022_unique, marker='o', color='red')
# ETICHETTE
plt.title('Evoluzione dei Casi Totali in Italia nel 2022', fontsize=14)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Totale Casi', fontsize=12)
# FORMATTAZIONE ASSE Y
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x/10000):,}K'))
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(1250000))
# FORMATTAZIONE ASSE X
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
# GRIGLIA DI VISUALIZZAZIONE
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.7, color='gray')
# OTTIMIZZAZIONE LAYOUT
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# MI ASSICURO CHE LE DATE SIANO ORDINATE
df_italy_2022_unique = df_italy_2022_unique.sort_values(by='date')

# GRAFICO
plt.figure(figsize=(18, 8))
sns.lineplot(x='date', y='new_cases', data=df_italy_2022_unique,color='grey')
plt.scatter(df_italy_2022_unique['date'], df_italy_2022_unique['new_cases'], color='red', zorder=5)
plt.title('Numero settimanale di nuovi casi in Italia (2022)')
plt.xlabel('Data', fontsize=12)
plt.ylabel('Nuovi Casi', fontsize=12)
plt.xticks(ticks=df_italy_2022_unique['date'], labels=df_italy_2022_unique['date'].dt.strftime('%Y-%m-%d'), rotation=45,fontsize=8)
plt.grid(True)
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x/1000)}k'))
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(50000))
plt.tight_layout()
plt.show()
```

```python
df_filtered_germany = df[(df['location'].isin([ 'Germany'])) & 
                 (df['date']>= '2023-01-01') & (df['date'] <= '2023-06-27')]
df_filtered_germany = df_filtered_germany.loc[:, ['location', 'date', 'hosp_patients','weekly_hosp_admissions']]
print(df_filtered_germany)
```

```python
sum_admissions = df_filtered_germany['weekly_hosp_admissions'].sum()  
num_days = df_filtered_germany['date'].nunique()
mean_admissions = df_filtered_germany['weekly_hosp_admissions'].mean()  
median_admissions = df_filtered_germany['weekly_hosp_admissions'].median() 
std_admissions = df_filtered_germany['weekly_hosp_admissions'].std()  

def format_number(num):
    return f"{num:,.0f}".replace(",", "X").replace(".", ",").replace("X", ".")
print("📊 Statistiche sui nuovi ricoveri settimanali in Germania fino al 27 giugno 2023:")
print(f"🔹 Somma totale: {format_number(sum_admissions)}")
print(f"🔹 Numero di giorni analizzati: {format_number(num_days)}")
print(f"🔹 Media: {format_number(mean_admissions)}")
print(f"🔹 Mediana: {format_number(median_admissions)}")
print(f"🔹 Deviazione standard: {format_number(std_admissions)}")
```
